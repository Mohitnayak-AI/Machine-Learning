```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['target'] = iris.target
df
```

```python
x = df.drop('target',axis=1)
y = df.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
# Desision Tree
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier()
```

```python
tree_model.fit(x_train,y_train)
```

```python
# visualize the tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree_model,filled=True)
```

```python
# let set the max depth to 2
tree_model = DecisionTreeClassifier(max_depth=2)
```

```python
tree_model.fit(x_train,y_train)

# visualize the tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree_model,filled=True)
```

```python
y_pred = tree_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

<h2> Prepruning & Hyperparameter Tuning </h2>

```python
desision_tree_param = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features': ['auto','sqrt', 'log2']
}
```

```python
desision_tree_param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
tree_model = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(tree_model,param_grid=desision_tree_param,cv=5,scoring='accuracy')
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.best_estimator_
```

```python
y_pared = grid.predict(x_test)

cm = confusion_matrix(y_test,y_pared)
print(cm)
print(classification_report(y_test,y_pared))
```
