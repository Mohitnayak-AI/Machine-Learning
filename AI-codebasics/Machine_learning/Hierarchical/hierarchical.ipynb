## Agglomerative Clustering

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import cluster,datasets
from sklearn.preprocessing import StandardScaler
```

```python
n_sample = 1500
x,y = datasets.make_circles(
    n_samples=n_sample,
    factor=0.5,
    noise=0.05,
    random_state=42
)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x[:,0],x[:,1])
plt.title('Noise circles')
plt.xlabel('feature 1')
plt.ylabel('feature 2')
```

```python
from sklearn.cluster import KMeans,AgglomerativeClustering

km = KMeans(n_clusters=2,n_init='auto')
km.fit(x,y)

import seaborn as sns

sns.scatterplot(x=x[:,0],y=x[:,1],hue=km.labels_)
```

```python
aggkm = AgglomerativeClustering(n_clusters=2,linkage='single')
aggkm.fit(x,y)

import seaborn as sns

sns.scatterplot(x=x[:,0],y=x[:,1],hue=aggkm.labels_)
```
